## Crear Tabla

```python
#Definir columnas
titulo = str

file.write('<table class="table table-responsive" id="myTable">\n\
  <thead>\n\
  <tr>\n\
    <th scope="col-2">Tipo de archivo</th>\n\
    <th scope="col-4">Descripción</th>\n\
    <th scope="col-4">Acciones</th>\n\
  </tr>\n\
  </thead>\n\
  <tbody>\n')
#Agregar datos
for indice in  range(len(datos[0])):  
    filename = datos[0][indice]
    file.write('<!---2021---> \n \
          <tr>\n \
            <td class="icon-file" data-label="Tipo de archivo">\n \
              <img src="IMG/images/ico-filekind-pdf.svg">\n \
            </td>\n')
    #Cambiar caracters desconocidos
    for palabra in filename:
            if(palabra == "#"):
                filename=filename.split("#")
                filename=" ".join(filename)
            if(palabra == "_"):
                filename=filename.split("_")
                filename=" ".join(filename)
                
    file.write(f'<td class="description" data-label="Descripción">\n \
              <strong class="law-title">{filename.lower()} ({filename[-3:]}-{pesos[indice]}KB)</strong>\n \
              <p class="date-descrip">Jueves 9 de septiembre de 2021 :</p>\n \
            </td>\n \
            <td class="buttons-section" data-label="Acciones">\n \
              <ul class="list-actions">\n')
        #Cambiar caracters espacios
    for palabra in filename:
            if(palabra == " "):
                filename=filename.split(" ")
                filename="_".join(filename)
                
    file.write(f'<li class="action-container">\n \
                  <a href="../IMG/{filename[-3:]}/{filename.lower()}" target="_blank"><button  class="btn btn-send btn-action see-btn" ></button></a>\n \
                  <p class="text-button">Ver</p>\n \
                </li>\n \
                <li class="action-container">\n \
                  <a href="../IMG/{filename[-3:]}/{filename.lower()}"  download=""><button  class="btn btn-send btn-action dwn-btn"></button></a>\n \
                  <p class="text-button">Descargar</p>\n \
                </li>\n \
                <li class="action-container">\n \
                  <a href="mailto:?cc=&amp;subject=&amp;body=%20-%20%20-Enlace:%20www.registraduria.gov.co/IMG/{filename[-3:]}/{filename.lower()}"  target="_blank"><button  class="btn btn-send btn-action share-btn"></button></a>\n \
                  <p class="text-button">Compartir</p>\n \
                </li>\n \
              </ul>\n \
            </td>\n \
          </tr>\n')
    
file.write('</tbody>\n\
        </table>')
file.close()
```
